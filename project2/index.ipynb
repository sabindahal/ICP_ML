```python
import time, math, json, os, random
from pathlib import Path
from dataclasses import dataclass
import numpy as np
import pandas as pd
from joblib import dump, load
from concurrent.futures import ThreadPoolExecutor, as_completed

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import matplotlib.pyplot as plt

# ====== Optional MinIO upload toggle ======
UPLOAD_TO_MINIO = False  # set True to auto-upload after saving
MINIO_ENDPOINT = os.getenv("MINIO_ENDPOINT", "127.0.0.1:9000")
MINIO_USER     = os.getenv("MINIO_USER", "minio")
MINIO_PASS     = os.getenv("MINIO_PASS", "minio123")
MINIO_BUCKET   = os.getenv("REGISTRY_BUCKET", "model-registry")

np.random.seed(42)
random.seed(42)

# =========================
# 1) Generate a synthetic "house price" dataset
# =========================
def make_synth_housing(n=20000):
    beds = np.random.randint(1, 6, size=n)
    baths = np.random.randint(1, 4, size=n)
    sqft = np.random.normal(1800, 600, size=n).clip(400, 5000)
    age  = np.random.exponential(20, size=n).clip(0, 120)
    dist = np.random.gamma(2., 3., size=n).clip(0, 40)
    garage = np.random.binomial(1, 0.6, size=n)
    school = np.random.normal(7.5, 1.5, size=n).clip(1, 10)

    price = (
        50000
        + beds * 12000
        + baths * 15000
        + sqft * 180
        - age  * 400
        - dist * 2000
        + garage * 8000
        + school * 5000
        + np.random.normal(0, 25000, size=n)
    )

    X = pd.DataFrame({
        "beds": beds,
        "baths": baths,
        "sqft": sqft.astype(int),
        "age": age,
        "dist_to_center_km": dist,
        "has_garage": garage,
        "school_score": school
    })
    y = price
    return X, y

X, y = make_synth_housing(20000)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=777)

num_features = X.columns.tolist()
preproc = ColumnTransformer(
    transformers=[("num", StandardScaler(), num_features)],
    remainder="drop"
)

# =========================
# 2) Define model variants
# =========================
@dataclass
class Variant:
    name: str
    model: object
    params: dict

def variants():
    return [
        Variant("Linear_v1", LinearRegression(), {}),
        Variant("Ridge_v1", Ridge(alpha=0.5), {"alpha": 0.5}),
        Variant("Ridge_v2", Ridge(alpha=5.0), {"alpha": 5.0}),
        Variant("Lasso_v1", Lasso(alpha=0.0005, max_iter=5000), {"alpha": 0.0005}),
        Variant("RF_v1", RandomForestRegressor(n_estimators=100, max_depth=12, n_jobs=-1, random_state=42), {"n_estimators":100,"max_depth":12}),
        Variant("RF_v2", RandomForestRegressor(n_estimators=300, max_depth=None, n_jobs=-1, random_state=42), {"n_estimators":300,"max_depth":None}),
        Variant("GB_v1", GradientBoostingRegressor(n_estimators=200, max_depth=3, random_state=42), {"n_estimators":200,"max_depth":3}),
        Variant("GB_v2", GradientBoostingRegressor(n_estimators=400, max_depth=3, random_state=42), {"n_estimators":400,"max_depth":3}),
    ]

# =========================
# 3) Train, evaluate, and persist variants (organized folders)
# =========================
REGISTRY_DIR = Path("local_model_registry")
REGISTRY_DIR.mkdir(exist_ok=True, parents=True)

def save_variant_artifacts(variant_name, pipeline, meta):
    """Save to local_model_registry/<variant>/{model.joblib, meta.json}"""
    vdir = REGISTRY_DIR / variant_name
    vdir.mkdir(parents=True, exist_ok=True)

    model_path = vdir / "model.joblib"
    dump(pipeline, model_path)

    meta_path = vdir / "meta.json"
    with open(meta_path, "w") as f:
        json.dump(meta, f, indent=2)
    return model_path, meta_path

def upload_folder_to_minio(local_dir: Path, bucket: str):
    from minio import Minio
    client = Minio(MINIO_ENDPOINT, access_key=MINIO_USER, secret_key=MINIO_PASS, secure=False)
    if not client.bucket_exists(bucket):
        client.make_bucket(bucket)
    for p in local_dir.rglob("*"):
        if p.is_file():
            object_name = f"{p.relative_to(local_dir).as_posix()}"
            client.fput_object(bucket, object_name, str(p))
    print(f"✅ Uploaded folder to MinIO bucket '{bucket}': {local_dir}")

def train_and_save(variant: Variant):
    pipe = Pipeline([
        ("prep", preproc),
        ("model", variant.model)
    ])
    t0 = time.perf_counter()
    pipe.fit(X_train, y_train)
    train_time = (time.perf_counter() - t0) * 1000.0  # ms

    # Evaluate
    y_pred = pipe.predict(X_test)
    rmse = math.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    meta = {
        "variant": variant.name,
        "params": variant.params,
        "train_ms": round(train_time, 2),
        "rmse": round(rmse, 3),
        "r2": round(r2, 4),
        "features": num_features,
    }
    model_path, meta_path = save_variant_artifacts(variant.name, pipe, meta)
    return meta, model_path, meta_path

all_meta = []
for v in variants():
    meta, _, _ = train_and_save(v)
    all_meta.append(meta)

df_meta = pd.DataFrame(all_meta).sort_values(by="rmse")
print("Base accuracy results:")
display(df_meta)

# Optional: upload the whole organized registry to MinIO
if UPLOAD_TO_MINIO:
    upload_folder_to_minio(REGISTRY_DIR, MINIO_BUCKET)

# =========================
# 4) Measure per-step latency (prep, infer, post) from organized layout
# =========================
def model_paths(variant_name: str):
    vdir = REGISTRY_DIR / variant_name
    return vdir / "model.joblib", vdir / "meta.json"

def measure_per_step_latency(variant_name: str, batch_size=256, iters=50):
    model_path, _ = model_paths(variant_name)
    pipe = load(model_path)
    prep = pipe.named_steps["prep"]
    core = pipe.named_steps["model"]

    Xb = X_test.sample(n=min(batch_size, len(X_test)), random_state=123)

    prep_times, infer_times, post_times = [], [], []

    def postproc(y):
        return np.round(y, 0)

    for _ in range(iters):
        t0 = time.perf_counter()
        Xp = prep.transform(Xb)      # preprocess only
        t1 = time.perf_counter()
        yhat = core.predict(Xp)      # model infer only
        t2 = time.perf_counter()
        _ = postproc(yhat)           # postproc only
        t3 = time.perf_counter()

        prep_times.append((t1 - t0) * 1000)
        infer_times.append((t2 - t1) * 1000)
        post_times.append((t3 - t2) * 1000)

    return {
        "prep_ms_p50": np.percentile(prep_times, 50),
        "infer_ms_p50": np.percentile(infer_times, 50),
        "post_ms_p50": np.percentile(post_times, 50),
        "total_ms_p50": np.percentile(np.array(prep_times) + np.array(infer_times) + np.array(post_times), 50),
        "prep_ms_p95": np.percentile(prep_times, 95),
        "infer_ms_p95": np.percentile(infer_times, 95),
        "post_ms_p95": np.percentile(post_times, 95),
        "total_ms_p95": np.percentile(np.array(prep_times) + np.array(infer_times) + np.array(post_times), 95),
    }

latency_rows = []
for v in variants():
    row = measure_per_step_latency(v.name, batch_size=512, iters=60)
    row["variant"] = v.name
    latency_rows.append(row)

df_latency = pd.DataFrame(latency_rows).sort_values("total_ms_p50")
print("Per-step latency (ms):")
display(df_latency[["variant","prep_ms_p50","infer_ms_p50","post_ms_p50","total_ms_p50","total_ms_p95"]])

# =========================
# 5) Throughput & scaling simulator (replicas × cores)
# =========================
SLA_LATENCY_MS = 500.0
SLA_RPS = 20.0

def single_request(pipe, Xb):
    t0 = time.perf_counter()
    yhat = pipe.predict(Xb)
    yhat = np.round(yhat, 0)
    dt = (time.perf_counter() - t0) * 1000.0
    return dt

def run_load_test(variant_name: str, reqs=200, batch_size=128, replicas=1, cores_per_replica=1):
    model_path, _ = model_paths(variant_name)
    pipe = load(model_path)

    Xbs = [X_test.sample(n=min(batch_size, len(X_test)), replace=True) for _ in range(reqs)]

    concurrency = max(1, replicas * cores_per_replica)
    latencies = []

    t_wall_start = time.perf_counter()
    with ThreadPoolExecutor(max_workers=concurrency) as ex:
        futures = [ex.submit(single_request, pipe, Xbs[i]) for i in range(reqs)]
        for fut in as_completed(futures):
            latencies.append(fut.result())
    wall_s = time.perf_counter() - t_wall_start

    rps = reqs / max(wall_s, 1e-9)

    metrics = {
        "p50_ms": float(np.percentile(latencies, 50)),
        "p95_ms": float(np.percentile(latencies, 95)),
        "rps": float(rps),
    }
    return metrics

def grid_scale_eval(variant_names, replicas_list, cores_list, reqs=200, batch_size=128):
    rows = []
    for vn in variant_names:
        _, meta_path = model_paths(vn)
        with open(meta_path) as f:
            meta = json.load(f)
        rmse = meta["rmse"]

        for r in replicas_list:
            for c in cores_list:
                m = run_load_test(vn, reqs=reqs, batch_size=batch_size, replicas=r, cores_per_replica=c)
                rows.append({
                    "Variant": vn,
                    "Scale": f"{r}×{c}",
                    "Replicas": r,
                    "Cores": c,
                    "Latency_p50_ms": round(m["p50_ms"], 2),
                    "Latency_p95_ms": round(m["p95_ms"], 2),
                    "Throughput_RPS": round(m["rps"], 2),
                    "Cost": f"{r}×{c}",
                    "RMSE": rmse
                })
    return pd.DataFrame(rows)

variant_names = [v.name for v in variants()][:6]  # adjust if you want all
replicas_list = [1, 2, 4]
cores_list    = [1, 2, 4, 8]

df_scale = grid_scale_eval(variant_names, replicas_list, cores_list, reqs=300, batch_size=256)
print("Scale sweep (latency/throughput/cost/accuracy):")
display(df_scale.head(20))

# =========================
# 6) SLA selection
# =========================
def choose_configs_meeting_sla(df, sla_latency_ms=SLA_LATENCY_MS, sla_rps=SLA_RPS):
    return df[(df["Latency_p95_ms"] <= sla_latency_ms) & (df["Throughput_RPS"] >= sla_rps)]

df_meet = choose_configs_meeting_sla(df_scale)
print(f"Configs that meet SLA (<= {SLA_LATENCY_MS} ms p95 and >= {SLA_RPS} RPS):")
display(df_meet.sort_values(["RMSE","Latency_p95_ms","Throughput_RPS"]).head(15))

def pick_optimal(df_meet):
    if df_meet.empty:
        return None
    dfc = df_meet.copy()
    dfc["TotalCores"] = dfc["Replicas"] * dfc["Cores"]
    dfc = dfc.sort_values(["RMSE", "TotalCores", "Latency_p95_ms", "Throughput_RPS"])
    return dfc.iloc[0]

optimal = pick_optimal(df_meet)
print("Suggested configuration (demo heuristic):")
print(optimal if optimal is not None else "No config meets SLA. Consider more replicas/cores or a faster variant.")

# =========================
# 7) Plots: accuracy vs latency, accuracy vs throughput
# =========================
agg = (df_scale
       .groupby("Variant", as_index=False)
       .agg(RMSE=("RMSE","mean"), Latency_p95_ms=("Latency_p95_ms","median"),
            Throughput_RPS=("Throughput_RPS","median")))

plt.figure(figsize=(8,5))
plt.scatter(agg["Latency_p95_ms"], agg["RMSE"])
for _, row in agg.iterrows():
    plt.annotate(row["Variant"], (row["Latency_p95_ms"], row["RMSE"]), xytext=(5,5), textcoords='offset points')
plt.xlabel("Latency p95 (ms)"); plt.ylabel("RMSE (↓ better)")
plt.title("Model Variant: Accuracy vs Latency"); plt.grid(True); plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(agg["Throughput_RPS"], agg["RMSE"])
for _, row in agg.iterrows():
    plt.annotate(row["Variant"], (row["Throughput_RPS"], row["RMSE"]), xytext=(5,5), textcoords='offset points')
plt.xlabel("Throughput (RPS, ↑ better)"); plt.ylabel("RMSE (↓ better)")
plt.title("Model Variant: Accuracy vs Throughput"); plt.grid(True); plt.tight_layout()
plt.show()

# =========================
# 8) Milestone Table
# =========================
report_cols = ["Variant","Scale","Latency_p95_ms","Cost","Throughput_RPS","RMSE"]
df_report = df_scale[report_cols].copy().sort_values(["Variant","Latency_p95_ms"])
print("Milestone-style Report Table:")
display(df_report.head(30))
```
